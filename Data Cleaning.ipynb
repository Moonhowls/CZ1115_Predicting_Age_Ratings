# Data Preparation

Okay, with all our data in json files, we are ready for our next step, data preparation.

The data are not be in a format that we like or are not usable, so we have to prepare the data for our use and extract the relevant information. 



```python
import pandas as pd
import json
import random
```

## Formatting data and convert into dataframe

First, we import in the data from our movie detail json file.

```python
def read_json(filename: str) -> dict:
  with open(filename, 'r') as f:
    return json.load(f)
```

### 1. Cleaning Data

In our code below, we clean up the data, placing different variables in separate lists.

When cleaning up our data, we only extract out the variables that are relevant.

```python
def cleaning_data (movie_detail):
    clean_data = {'id':[],'runtime':[],'genres':[],'keywords':[],'production_companies':[],'age_rating':[],
                  'budget':[]}
    i=0
    for movie in movie_detail:
        i+=1
        print("Cleaning", i , "-th movie")
        st = False
        
        if movie != None:
            for certification in movie["release_dates"]['results']:
                
                if (certification["iso_3166_1"] == 'US'):
                    st = True
                    clean_data['age_rating'].append(certification['release_dates'][0]["certification"])
                    break
            
            if st==False:
                # This is when the data does not have US rating, so we have to ignore the data
                continue

            # id
            clean_data['id'].append(movie['id'])

            # runtime
            clean_data['runtime'].append(movie['runtime'])

            #genres
            genres_list = []
            for genre in movie['genres']:
                genres_list.append(genre['name'])
            clean_data['genres'].append(genres_list)
            
            #Keywords
            keywords_list = []
            for keywords in movie['keywords']['keywords']:
                keywords_list.append(keywords['name'])
            clean_data['keywords'].append(keywords_list)
            production_companies = []
            for production in movie['production_companies']:
                production_companies.append(production['name'])
            clean_data['production_companies'].append(production_companies)

            #budget
            clean_data['budget'].append(movie['budget'])

            
    return clean_data
```

```python
movie_detail = read_json("movie_detail.json")
print("The length of movie_detail is ", len(movie_detail))
```

```python
clean_data = cleaning_data(movie_detail)
```

### 2. Converting into DataFrame

After cleaning our data, we create a DataFrame for the clean data.

This will allow us to use and manipulate the data as we wish.

```python
df = pd.DataFrame(clean_data)
df['age_rating'].astype("category")
print("The shape of the dataFrame is" , df.shape)
df.head()
```

Extract data that is only in the 5 interest Age_rating (refer to the data exploration - there is NR)

```python
df = df [ df['age_rating'].isin(['PG','PG-13','R','NC-17','G'])]
print("The shape of the dataFrame is" , df.shape)
df
```

There are 10000 movie data initial, but after running to get the movie_Details, there are only 9038 data that is valid (have US age_rating and also not empty data) left. However some of the data contains age_rating NR (which is No Rating and we ignore this data as stated in Data Exploration), we left 6596 data

```python
df.to_csv("movie_data.csv", header=True,index=False)
```
